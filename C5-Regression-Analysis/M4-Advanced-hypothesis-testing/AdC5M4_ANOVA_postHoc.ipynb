ANOVA post hoc tests with Python

```python
import seaborn as sns
import statsmodels.api as sm
from statsmodels.formula.api import ols
from statsmodels.stats.multicomp import pairwise_tukeyhsd
```

```python
diamonds = sns.load_dataset("diamonds")
```

```python
# Fit OLS model
model = ols(formula = "price ~ C(color)", data = diamonds).fit()
```

```python
# Perform ANOVA
sm.stats.anova_lm(model, type = 2)
```

 The p-values are much smaller than 0.05 so the results are statistically significant and therefore, it can ce rejected that the null hypothesis which the mean price is the same across all diamond colors.

```python
# Perform Tukey's HSD test to compare the mean prices of diamonds across different colors
tukey_oneway = pairwise_tukeyhsd(endog = diamonds["price"], groups = diamonds["color"], alpha = 0.05)
tukey_oneway.summary()
```

If the reject column reads false, the null hypothesis cannot be rejected. If the reject column reads true, the null hypothesis can be rejected.

The null hypothesis is that the diamond prices between the two color grades are the same. But the Tukey's HSD test informs that the null hypothesis with a p-value of 0.001 can be rejected. For example, the diamond price is not the same for H and I grade diamonds.
