Confidence intervals with Python

```python
import numpy as np
import pandas as pd
from scipy import stats
```

Code to have python simulate the same random sample of district literacy rate data.

```python
education_districtwise = pd.read_csv("/content/education_districtwise.csv")
education_districtwise = education_districtwise.dropna()
sampled_data = education_districtwise.sample(n=50, replace=True, random_state=31208)
```

Compute the mean district literacy rate for your sample data.

```python
sample_mean = sampled_data['OVERALL_LI'].mean()
sample_mean
```

Standard Error of the mean

$SE$($\bar{x}$) = $\frac{s}{\sqrt{n}}$

```python
estimated_standard_error = sampled_data['OVERALL_LI'].std() / np.sqrt(sampled_data.shape[0])
```

Construct confidence interval with the stats.norm.interval function. First, write out the function and set the arguments. For alpha, set 0.95, because you want to use a 95% confidence level. For loc enter the variable sample_mean. And for scale, enter the variable, estimatimated_standard_error.

```python
stats.norm.interval(confidence=0.95, loc=sample_mean, scale=estimated_standard_error)
```

 A 95% confidence interval for the mean district literacy rate, that stretches from about 71.4% to 77.0%.

 A senior director in the Department wants to be even more confident about your results. The director wants to make sure you have a reliable estimate, and suggest that you recalculate your interval with a 99% confidence level.

```python
stats.norm.interval(confidence=0.99, loc=sample_mean, scale=estimated_standard_error)
```

A 99% confidence interval for the mean district literacy rate that stretches from about 70.5% to 77.9%.

This results will help the Department of Education decide how to distribute government resources to improve literacy.
